# Gathering Data

```python
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["North Korea"]
```

```python
pytrend.get_historical_interest(kw_list, 
                                 year_start=2018, month_start=1, 
                                 day_start=1, hour_start=0, 
                                 year_end=2018, month_end=2, 
                                 day_end=1, hour_end=0, 
                                 cat=0, geo='', gprop='', sleep=10)
```
